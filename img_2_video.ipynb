```python
import glob
from moviepy.editor import *

for folder_path in glob.glob('../sber-swap/jtbc_output_img/*'):

    files=[]
    li = glob.glob(folder_path+'/*.jpg')
    li.sort()
    for i in li:
        files.append(i)

    folder_name = folder_path.split('/')[-1]
    clip = ImageSequenceClip(files, fps = 30)
    clip.write_videofile(f'../sber-swap/jtbc_output_mp4/{folder_name}.mp4')
```
